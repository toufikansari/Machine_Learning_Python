```python
# Import all Lybrary
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
import matplotlib.pyplot as plt
import numpy as np


#Load Data 
data = pd.read_csv("data.csv")

data.to_excel("data.xlsx")

# Input & Output

x = data[['Hours_Studied']]
y = data['Marks']


#train Model

model = LinearRegression()
model.fit(x , y)
predict_score = model.predict(x)

#Viled Regression Metrics
mae = mean_absolute_error(y , predict_score)
mse = mean_squared_error(y , predict_score)
rmse = np.sqrt(mse)
r2 = r2_score(y , predict_score)

#Show Result

"""
suppose your ans = 1.1234421
round(mse , 2) = it will take two value after point
like ----- (1.12)
"""
print("Mean absulate error :", round(mae , 2))
print("Mean square error :", round(mse , 2))
print('RAMSE :', round(rmse , 2))
print("r2_score (Model Accuracy)" ,round(r2 , 2) ) #Closer to 1 = better

#histogram
plt.Figure(figsize=(10, 6))
plt.hist(data["Marks"] ,bins=30 , color='blue' , edgecolor='black' )
plt.title('Distributan of Final Exam Score')
plt.xlabel('Final Exam score')
plt.ylabel('No Of Student')
plt.grid(True)
plt.show()

#Scater + Regression line 
plt.Figure(figsize=(10, 6))
plt.scatter(x , y, color='green', label='Actual Score'  )
plt.plot(x , predict_score , color='red' , label='Pridected Score Regression Line' )
plt.title('Model Prediction Vs Actual Score')
plt.xlabel('Study Hour')
plt.ylabel('Final OutPut')
plt.grid(True)
plt.show()

New_Hour = 9
predict_scorenew = model.predict([[New_Hour]])
print(f"Pridect final score for {New_Hour} hours is {predict_scorenew} ")
```
